```python
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
```

```python
from sklearn.metrics import davies_bouldin_score
```

```python
seed_values = [1,10,100,1000,1000]
for i in range (3,8) :
    for seeds in seed_values :
        np.random.seed(seeds)
        kmeans= KMeans(n_clusters=i)
        labels_kmeans = kmeans.fit_predict(df)
        db_score = davies_bouldin_score(df,labels_kmeans)
        print (" The davies bouldin score for seed",seeds, " and number of clusters",i," is :",db_score)
    

```

For KMeans good nb of clusters is 4 

```python
seed_values = [1,10,100,1000,1000]
for i in range (3,8) :
    for seeds in seed_values :
        np.random.seed(seeds)
        h_cluster= AgglomerativeClustering(n_clusters=i)
        h_labels = h_cluster.fit_predict(df)
        db_score = davies_bouldin_score(df,h_labels)
        print (" The davies bouldin score for seed",seeds, " and number of clusters",i," is :",db_score)
```

```python
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
```

```python
neighbors = NearestNeighbors(n_neighbors=2)
neighbors_fit = neighbors.fit(df)
distances, indices = neighbors_fit.kneighbors(df)
distances = np.sort(distances, axis=0)
distances = distances[:,1]
```

```python
plt.plot(distances)
```

```python
plt.plot(distances)
plt.xlim(2500,2850)
plt.show()
```

it seems like eps=0.3

Davies bouldin is not a good metric for dbscan's purpose

```python
eps = 0.25
minPts = 23
```

```python
db_clustering= DBSCAN(eps=eps,min_samples=minPts)
labels_dbscan = db_clustering.fit_predict(df)
df['labels_dbscan'] = labels_dbscan
plt.figure(figsize=(12,8))
sns.scatterplot(x='V1',y='V2',data=df, hue= 'labels_dbscan',palette='tab10')
```
